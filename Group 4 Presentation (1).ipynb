> # 1. How can Epsilon go to zero?

Working with complex refractive index and dielectric constant we have:
$$
n = n_r + i n_i \\
ϵ = ϵ_r + i ϵ_i
$$

using the same fundamental relation $n^2 = \epsilon_0 ϵ$

$$
n_r = \sqrt{\frac{| \epsilon| + \epsilon_r}{2 \epsilon_0}}\\
n_i = \sqrt{\frac{| \epsilon| - \epsilon_r}{2 \epsilon_0}}
$$

These relations will be useful for us later on when we write things out as linear and non linear terms.

$$
D = ϵ_0 E + P = \epsilon_0 (1 + \chi) E = \epsilon_0 (1 + \chi^{(1)}(w) + 3 \chi^{(3)} (w)|E|^2) E = \epsilon E\\
\epsilon = \epsilon^{(1)} + \epsilon^{(NL)}\\
\implies \epsilon^{(NL)}_r = 3 \epsilon_0 \chi_r^{(3)} |E|^2
, \epsilon^{(NL)}_i = 3 \epsilon_0 \chi_i^{(3)} |E|^2
$$
similary we write $n = n^{(1)} + n^{(NL)} = n^{(1)} + n^{(2)} I$ (intensity dependent refractive index, $I = 2 \epsilon_0 n^{(1)} c |E|^2$). Using all the relations above we simplify to obtain refractive index as a function of susceptibiity:
$$
n^{(2)}_r = \frac{3}{4\epsilon_0 c( (n_r^{(1)})^2 + (n_i^{(1)})^2 }[ \chi_r^{(3)} + \frac{n^{(1)}_i}{n^{(1)}_r} \chi_i^{(3)}] \\
n^{(2)}_i = \frac{3}{4\epsilon_0 c( (n_r^{(1)})^2 + (n_i^{(1)})^2 }[ \chi_r^{(3)} - \frac{n^{(1)}_i}{n^{(1)}_r} \chi_i^{(3)}] 
$$
having derived these relations we can now describe what we mean by ENZ. For a material with complex dielectric constant (which is dependent on the frequency $ \omega$) for a particular $\omega$, $\epsilon^{(1)}_r = 0$ while imaginary part remains nonzero.


> # 2. Two Level Systems

In Density matrix formalism, Schrodinger's equation takes the form of Neumann's equation:

$$\dot{\rho} = \frac{-i}{\hbar}[\hat{H}, \hat{\rho}]$$

The Hamiltonian for a two-level atomic system is given by $\hat{H} = \hat{H}_0 + \hat{V}(t)$, where $\hat{H}_0$ is the default Hamiltonian of the system and $\hat{V}(t)$ represents the interaction of the atomic system with electromagnetic radiation. <break>

$$
 \hat{H}_{nm} = \delta_{nm}E_n+ \hat{V}_{nm}
$$

We can now write the density matrix equations of this system , by substituting $H$ , for each of the diagonal and off-diagonal and introduce phenomenological time constants $T_1$ (relaxation time to deacy from b to a) and $T_2$ (dipole dephasing time) to get:

$$
\begin{aligned}
\dot{\rho}_{aa} &= \frac{\rho_{bb}}{T_1} - \frac{i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ba}}) \\
\dot{\rho}_{bb} &= -\frac{\rho_{bb}}{T_1} + \frac{i}{\hbar}({V_{ab}{\rho}_{ba} - {\rho}_{ab}{V}_{ba}}) \\
\dot{\rho}_{ab} &= \frac{i}{h}V_{ab}(\rho_{aa} - \rho_{bb})- (i\omega_{ab} + \gamma_{ab})\rho_{ab}
\end{aligned}
$$
 Now using properties of a density matrix, that $\rho = \rho^{\dagger}$ and $Tr(\rho) = 1$, we can reduce these to :


$$
\begin{align}
\tag{1}
\boxed{\frac{d{\rho}_{ab}}{dt} = \frac{i}{h}V_{ab}\Delta{\rho}- (i\omega_{ab} + \gamma_{ab})\rho_{ab}}
\end{align}
$$
    
    
$$
\begin{align}
\tag{2}
\boxed{\frac{d\Delta{\rho}}{dt} = \frac{-(\Delta{\rho} - 1)}{T_1} + \frac{2i\mu_{21}E(t)}{\hbar}({\rho}_{ba} - {\rho}_{ab})}
\end{align}
$$

where, $\Delta \rho = \rho_{aa}-\rho_{bb}$    
    
    
From these density matrix equations, we write the electric field as $E(t) = E_1e^{-i{\omega_1}t} + E_2e^{-i{\omega_2}t} + c.c$. Under the rotating wave approximation, we will discard terms that have an exponent of $\omega_1 + \omega_2$. We proceed to use perturbation theory, and write $\Delta{\rho}$ and $\rho_{21}$ as

$$
\begin{align}
\tag{3}
\Delta{\rho} = 1 - \delta_{12}e^{-i({\omega_1 - \omega_2})t} - \delta_{12}^{*}e^{i({\omega_1 - \omega_2})t} + ... 
\end{align}
$$
$$
\begin{align}
\tag{4}
\rho_{21} = {{\sigma}_1}{e^{-i{\omega_1}t}} + {{\sigma}_2}{e^{-i{\omega_2}t}} + {{\sigma}_{112}}e^{-i({2\omega_1 - \omega_2})t} + {{\sigma}_{211}}e^{-i({2\omega_2 - \omega_1})t} + ... 
\end{align}
$$

We need to calculate the third order susceptibility. So we proceed by calculating all the coefficients of the perturbative expansion till $\sigma_{112}$. 

$$
\begin{aligned}
\sigma_{1(2)} &= \frac{{\mu_{21}}E_{1(2)}{\hbar}^{-1}}{({\omega_{21}} - \omega_{1(2)}) - i\gamma}
\end{aligned}
$$

$$
\begin{aligned}
\delta_{12} &= \frac{{\omega_1 - \omega_2 + 2i\gamma}}{(\omega_1 - \omega_2 + \frac{i}{T_1})}(\frac{{\mu_{21}^{2}E_1E_2}/{\hbar^{2}}}{({\omega_{21}} - \omega_{1} - i\gamma)({\omega_{21}} - \omega_{2} + i\gamma)})
\end{aligned}
$$

$$
\begin{aligned}
{{\sigma}_{112}} &= \frac{{\mu_{21}}\delta_{12}E_1/{\hbar}}{(\omega_{2} + \omega_{21} -2\omega_1  - i\gamma)} = \frac{{\omega_1 - \omega_2 + 2i\gamma}}{(\omega_1 - \omega_2 + \frac{i}{T_1})}(\frac{{\mu_{21}^{3}E_1^2E_2}/{\hbar^{3}}}{({\omega_{21}} - \omega_{1} - i\gamma)({\omega_{21}} - \omega_{2} + i\gamma)(\omega_{2} + \omega_{21} -2\omega_1  - i\gamma)})
\end{aligned}
$$

Now, we can obtain the third order susceptibility $\chi^{(3)}$ since $P^{(3)}(2\omega_1 - \omega_2) = N\mu_{21}\sigma_{112}e^{-i(2\omega_1 - \omega_2)t} = \epsilon_0\chi^{(3)}E_1^{2}E_2e^{-i(2\omega_1 - \omega_2)t}$, where $N$ is the number density of the two-level system. We make the assumption that $\omega_1$ and $\omega_2$ are sufficiently close that $\Delta{\omega}<<\omega_1, \omega_2$. Therefore, we can approximate both $\omega_1$ and $\omega_2$ by $\bar{\omega} = \frac{\omega_1 + \omega_2}{2}$. Hence,

$$
\begin{aligned}
\chi^{(3)} &= \frac{N\mu_{21}\sigma_{112}}{\epsilon_0E_1^{2}E_2} \\
\chi^{(3)} &\approx \frac{{\Delta\omega + 2i\gamma}}{(\Delta\omega + \frac{i}{T_1})}(\frac{{N\mu_{21}^{4}}/{\epsilon_{0}\hbar^{3}}}{({(\omega_{21}} - \bar{\omega})^2 + \gamma^2)^{\frac{3}{2}}}) = (1 + \frac{{2\frac{T_1}{T_2} - 1}}{1 -i\Delta{\omega}T_1})(\frac{{N\mu_{21}^{4}}/{\epsilon_{0}\hbar^{3}}}{({(\omega_{21}} - \bar{\omega})^2 + \gamma^2)^{\frac{3}{2}}}) = \chi_{fast}^{(3)} +  \chi_{slow}^{(3)} 
\end{aligned}
$$




```python
def sus(delw,t1,t2):
  return abs(1 + (2*t1/t2 - 1)/(1 - delw * t1 * 1j))
def susfast(delw,t1,t2):
  return abs((2 * t1/t2 - 1)/(1 - delw * t1 * 1j))

plt.figure(figsize=(10, 6))
delws = np.logspace(0, 14, 100)
t1, t2 = 1e-9, 1e-12
sussies = sus(delws, t1, t2)
norm = sussies[0]
sussies /= norm

plt.title('Fast and Slow Susceptibilities vs Beats Frequency', fontsize=20)
plt.plot(delws, sussies)
plt.plot(delws, susfast(delws, t1, t2)/norm, c='r', ls='--', label='Fast', lw=2)
plt.axhline(1/norm, c='b', ls='--', label='Slow', lw=2)
plt.axvline(1/t1, c='k', ls='--', label=r'$T1^{-1}$')
plt.axvline(1/t2, c='k', ls='--', label=r'$T2^{-1}$')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Susceptibility (arbitrary units)', fontsize=15)
plt.xlabel('Beats frequency (Hz)', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=14)
plt.show()
```

### Why call them fast and slow ?

$$\boxed{\chi^{(3)}_{slow} = (\frac{{2\frac{T_1}{T_2} - 1}}{1 -i\Delta{\omega}T_1}) \frac{{N\mu_{21}^{4}}/{\epsilon_{0}\hbar^{3}}}{({(\omega_{21}} - \bar{\omega})^2 + \gamma^2)^{\frac{3}{2}}}}$$

$$\boxed{\chi^{(3)}_{fast}=\frac{{N\mu_{21}^{4}}/{\epsilon_{0}\hbar^{3}}}{({(\omega_{21}} - \bar{\omega})^2 + \gamma^2)^{\frac{3}{2}}}}$$

$$\boxed{ \frac{\chi^{(3)_{slow}}}{\chi^{(3)_{fast}}} =  (\frac{{2\frac{T_1}{T_2} - 1}}{1 -i\Delta{\omega}T_1}) }$$

At resonant frequency, real absorption of photons takes place which takes more time than excitation to virtual states, leading to fast and slow processes! Since $T_2 << T_1$, the slow term dominates at resonance

> ## Periodic Kronig-Penny model and The Bloch Theorem
If we ignore edge effects, we can write:
$$V(x+p) = V(x+a+b) = V(x)$$
In such potentials, the wave function must be of the form,
$$|\psi(x+p)|^2 = |\psi(x)|^2 \implies \psi(x+p) = e^{ikp}\psi(x)$$

$$e^{ikpN}=1 \equiv e^{i2n\pi}$$
$$k=\pm \frac{2\pi n}{Np}, n = -\frac{N}{2}, .., -1, 0, 1, .., \frac{N}{2}$$
$$\implies k_{max}=\frac{\pi}{p}, k_{min}=-\frac{\pi}{p}$$

The Bloch Theorem states that valid wavefunctions for such a periodic potential must take the form $$\psi(r)=e^{ikr}u(r)$$

**Applying boundary conditions, we get:**
$$\frac{1-2\xi}{2\sqrt{\xi(\xi-1)}}\sin\alpha_0a\sqrt{\xi}\sin\alpha_0b\sqrt{\xi-1}+\cos\alpha_0a\sqrt{\xi}\cos\alpha_0b\sqrt{\xi-1} = \cos k (a+b) \hspace{2em}\text{... when } \hspace{0.5em} E > U_0$$

Solving the above equations, we get the valid values of $E$, the energy of the particle. Notice that the $E$ can take only discrete values because $k$ can take only certain values. However, since $N$ is very large, the energy states can be said to be approximately continuous.

```python
# write the above equation as f(xi) = cos(k*(a+b))
# solve for xi to get the valid states of the system

def f(xi):
  return (1-2*xi)/(2*np.sqrt(xi*abs(1-xi)))*(np.sin(a*xi**0.5)*np.sin(a*abs(1-xi)**0.5) + np.cos(b*xi**0.5)*np.cos(b*abs(1-xi)**0.5))
 
kp = np.vectorize(f)
kpp = lambda xi: kp(xi)-1 # cos(k*(a+b)) = +1
kpm = lambda xi: kp(xi)+1 # cos(k*(a+b)) = -1

xis = np.linspace(1.01,2,1000)
a, b = 30, 30
fig, ax = plt.subplots(figsize=(10,8))

# solving for intersection points
minus1 = fsolve(kpm, [1.011,1.05,1.19,1.25,1.6])
plus1 = fsolve(kpp, [1.011,1.05,1.13,1.4,1.6])

# plotting
xranges = xis[np.logical_and(kp(xis)<1, kp(xis)>-1)]
plt.title('$f(\\xi)$ vs $\\xi$', fontsize=20)
plt.plot(xis,kp(xis), label='$f(\\xi)$')
plt.plot(xis, np.ones(len(xis)), label='$k=0$')
plt.plot(xis, -np.ones(len(xis)), label='$k=\\pm\\frac{\\pi}{a+b}$')
for i in range(5):
    ax.add_patch(Rectangle((min(minus1[i],plus1[i]),-1),abs(plus1[i]-minus1[i]),2,facecolor='orange',alpha = 0.7))
plt.ylabel(r'$f(\xi)$', fontsize=15)
plt.xlabel(r'$\xi$', fontsize=15);
plt.legend(fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

> ## Carrier Motion in Non-Parabolic Band

Consider the motion of electrons in a band with dispersion $E(k)$, in the presence of a harmonic electric field $E_z(t) = E_1e^{i\omega_1 t} + E_2e^{i\omega_2 t} + \text{c.c.}$

The equation of motion will thus be
$$\frac{dk_z}{dt} = -\frac{e}{\hbar}E_z - \gamma(k_z-k_{z0})$$ 
Where 
$\gamma$ is the momentum scattering rate. The solution for the harmonic input field will be $k_z(t) = k_{z0} + \delta k (t)$, where
$$ \boxed{\delta k(t) = \frac{e}{\hbar(i\omega_1-\gamma)}E^{-i\omega_1t} + \frac{e}{\hbar(i\omega_2-\gamma)}E^{-i\omega_2t} + \text{c.c.}}$$

The electron velocity in the band $E(k)$ in the direction of the applied electric field is $v_z = \hbar^{-1}\partial E/\partial k_z$, which can be expanded into the power series around $k_0$ as 
$$v_z(t) = v_z(k_{z0}) + \frac{\partial v_z}{\partial k_z} \delta k(t) + \frac{1}{2}\frac{\partial^2 v_z}{\partial k_z^2} \delta k^2(t) + \frac{1}{6}\frac{\partial^3 v_z}{\partial k_z^3} \delta k^3(t) + \cdots$$

The first remaining term that is linear in $\delta k(t)$ can be represented by introducing the _optical effective momentum dependent mass_
$$m_{t}^{-1} = \hbar^{-1} \frac{\partial v_z}{\partial k_z}$$
Averaging over all directions we get,
$$\boxed{\frac{1}{m_{t}}=\frac{1}{\hbar}\left[\frac{2}{3}\frac{v(k)}{k} +\frac{1}{3}\frac{dv(k)}{dk}\right] = \frac{1}{3\hbar k^2}\frac{d}{dk}(k^2v(k))}$$


> ## Kane's K.P. Model

From Kane's KP model we have
$$E(k) = \frac{E_{gap}}{2}\sqrt{1+\frac{2\hbar^2k^2}{m^*E_{gap}}}=\frac{E_{gap}}{2}\sqrt{1+(k/k_0)^2}$$
where $m^*$ is the effective mass at $k=0$, $E_{gap}$ is the bandgap energy and $k_0$ is the "non-parabolocity" wave vector given by
$$k_0 = \sqrt{\frac{m^*E_{gap}}{2\hbar^2}}$$
Note that the disperesion relation changes from "parabolic like" to "linear like" at $k=k_0$. 


```python
def E_non_parabolic(c):
  return np.sqrt(1+c**2)-1

plt.figure(figsize=(10, 8))
cs = np.linspace(-2, 2)
plt.plot(cs, E_non_parabolic(cs), label='Non Parabolic Energy')
plt.plot(cs, 0.5*cs**2, '--', label='Parabolic Approximation')
K = np.array([-1, 1])
plt.plot(K, E_non_parabolic(K), 'o', label='$|k|=k_0$')
plt.title('Energy vs $k_0$', fontsize=20)
plt.ylim(0, 1.5)
plt.xlabel('Wave vector ($k$ in units of $k_0$)', fontsize=15)
plt.ylabel('Energy (in units of $E_{gap}/2$)', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=15, loc='upper right')
plt.show()
```


Using the above relations we can determine the velocity
$$v(k) = \frac{1}{\hbar}\frac{\partial E}{\partial k} = v_{sat}\frac{k/k_0}{\sqrt{1+(k/k_0)^2}}$$
where the saturation velocity is 
$$v_{sat} = \frac{1}{2} \frac{E_{gap}}{\hbar k_0} = \frac{\hbar k_0}{m^*}$$

The momentum dependent optical mass can be calculated from the previously determined expression as:
$$m_t^{-1}=\frac{v_{sat}}{\hbar k_0}\frac{2(k/k_0)^2+3}{3[1+(k/k_0)^2]^{3/2}} = {m^*}^{-1}\frac{2(k/k_0)^2+3}{3[1+(k/k_0)^2]^{3/2}}$$

This perturbation changes v,
$$v(k) = v(k_{0}) + \frac{∂v}{∂k} ⃒ _{k_{0}} (\partial k(t))+ \frac{1}{2} \frac{∂^{2} v}{\partial k^{2}}(∂k^{2}(t))+ \frac{1}{6} \frac{∂^{3} v}{\partial k^{3}}(∂k^{3}(t)) + ...$$

Average v in k-space, use odd parity of $\alpha$,
$$<v> =\hbar[<m^{-1}_{t}> \partial k(t) + \frac{1}{6}<β>\partial^{3}k(t)  ] $$
where

 $ <\beta>= \frac{m^{-1}_{*}}{k_{0}^{2}}[ \frac{4 \sinh^{-1}(\frac{k_{F}}{k_{0}})}{(\frac{k_{F}}{k_{0}})^3}- \frac{4(\frac{k}{k_{0}})^4 + 11(\frac{k}{k_{0}})^2 +4 }{(\frac{k_{F}}{k_{0}})^2(1+ (\frac{k}{k_{0}})^2)^{5/2}} ]$
 



Find average $<x>$, relate it to $P_{NL}$ via induced dipole moment, since $v(\omega) = - i\omega x(\omega)$

$$\boxed{\chi^{(1)}(\omega_{1,2}) = \frac{N e^{2}<m_{t}^{-1}>}{ϵ (\omega_{1,2}+ i \gamma)}} $$

where,

$$f_{\beta}(k_{F})=k_{0}^{2}<\beta><m_{t}>$$

> ## 4.B Slow non-linearities in ENZ materials

$$\boxed{\epsilon_{r} = ϵ_{∞} - \frac{\omega_{p}}{\omega^{2}+i\omega\gamma} }$$

Where, 

$\omega_{p} = \sqrt{\frac{N e^{2}}{<m_{t}^{-1}>ϵ_{0}(\omega^{2}+ i \gamma \omega )}}$ , is the plasma frequency

$ϵ-{∞}$ is the permittivity at highly non-resonant frequency

$\omega$ is the drive frequency

$N$ is the carrier chare density in the conduction band

$\epsilon_{r}$ depends on $\omega_{p}$, which depends on $<m^{-1}_{t}>$, which depends on $\partial k(t)$; 


$$\delta P(t) = E(t) J(t)$$


Now, $J^{(1)}(t) = N e<v(t)^{(1)}> $


$$\boxed{J^{(1)}(\omega_{1,2})= i \frac{e^{2}N^{2}}{\omega_{1,2}+ i \gamma_{1,2}} E_{1,2}e^{-i \omega_{1,2}t}  + c.c }$$




$$\boxed{\delta P(t) = \frac{2Ne^{2}\gamma <m_{t}^{-1}>}{\bar\omega^{2}+\gamma^{2}}[E_{1}^{2} + E_{2}^{2} + 2 E_{1}E_{2}cos((\omega_{1}-\omega_{2})t)]}$$


To account for lattice relaxations:

$$\boxed{\frac{dU_{hot}(t)}{dt} = \delta P(t) - \frac{U_{hot}}{\tau_{el}}} $$


$$U_{hot}(t) = \frac{\tau_{el}}{1-iΔω\tau_{el}}\frac{2Ne^{2}\gamma <m_{t}^{-1}>}{\bar \omega^{2} + \gamma^{2}} E_{1}E_{2}e^{-i(\omega_{1}-\omega_{2})t} + c.c
$$

$$\delta E(t) = \frac{U_{hot}(t)}{f_{hot}N} $$

Now, $\delta E(t) $ relates $\partial k$, which relates $\partial m_{t}^{-1}$,


$$\delta \epsilon_{r}(t) ≈ \frac{\alpha_{F} U_{hot}(t) \chi^{(1)}(\bar \omega)}{<m_{t}^{-1}>N \hbar v_{F}} $$

in Fourier space,
$$\boxed{\delta ϵ(\omega_{1}-\omega_{2})≈ \alpha_{F}\frac{1}{N \hbar v_{F}}\frac{  \gamma\tau_{el}}{(1-iΔ\omega\tau_{el})} \frac{2N e^{2}}{(\bar\omega^{2}+ \gamma^{2})}\chi^{(1)}E_{1}E_{2}e^{-i(\omega_{1}-\omega_{2})t}}$$



A field $E_{1}e^{i\omega_{1}t}$ experiences this $\epsilon$, identify resulting field  with $P^{(3)}_{NL}$,

$$\boxed{\boxed{\chi^{(3)}_{slow}(2\omega_{1}-\omega_{2};\omega_{1};\omega_{2}) = \frac{ 2 (\tau_{el}/\tau_s)}{ (1-iΔ\omega\tau_{el})} \frac{ e^{2}\chi^{(1)} f_{\alpha}(k_{F})}{\hbar^{2}k_{0}^{2}(\bar\omega^{2}+ \gamma^{2})}}}$$

```python
def mt_avg(c):
  return (1+c**2)**0.5

def falpha(c):
  return (2*c**2+5)/(3*(1+c**2)**2)
  
def fbeta(c):
  return mt_avg(c) * beta_avg(c)

cs = np.linspace(1, 3, 1000)

#x = fsolve(lambda x: falpha(x) -, x0=1.5)
plt.figure(figsize=(10, 8))
plt.title('$f_\\alpha, f_\\beta$ vs $k_F/k_0$', fontsize=20)
plt.plot([1.6, 1.6], [-1, 1], '--', color='black', label='$x = 1.6$')
plt.plot([0, 4], [falpha(1.6), falpha(1.6)], '--', color='black')
plt.plot([0, 4], [fbeta(1.6), fbeta(1.6)], '--', color='black')
plt.xlim(0.9, 3.1)
plt.ylim(0.0, 0.7)
plt.plot(cs, falpha(cs), label='$f_{\\alpha}$')
plt.plot(cs, fbeta(cs), label='$f_{\\beta}$')
plt.ylabel('$f_{\\alpha}$, $f_{\\beta}$', fontsize=20)
plt.xlabel('$k_F/k_0$', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=15)
plt.show()
```

The value $\frac{k_{F}}{k_{0}} \approx 1.6$ is typical for most ENZ materials, from $f_{\beta} \approx 2 \times f_{\alpha}$

> # 5. Conclusions 

Our goal in this section is to compare fast and slow non linearities in "enz materials" (semiconductors operating at certain frequencies) to traditional 2 level systems exhitbiting fast and slow non linearities.

Figure of merit (FOM)—the product of maximum nonlinearity, propagation length, and the signal bandwidth B is useful to optimize materials. For ENZ materials it is given by 
$$
\text{maximum nonlinearity*propagation length*signal bandwidth B} = \frac{1}{2 \pi \tau_{el}}
$$ 
where $\tau_{el}$ is ENZ response time, which typically lies between 100fs and 1ps.

Similarly for traditional materials it comes out to be $\frac{1}{2 \pi T_1}$
